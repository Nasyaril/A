```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# membaca gambar
img = cv2.imread('/content/suisei.jpg', cv2.IMREAD_GRAYSCALE)

# fungsi kuantisasi
def quantization(image, levels):

    step = 256 // levels
    quantized = (image // step) * step

    return quantized

# kuantisasi
img_2bit = quantization(img, 4)     # 4 level
img_4bit = quantization(img, 16)    # 16 level
img_8bit = quantization(img, 256)   # 256 level

plt.figure(figsize=(10,6))

plt.subplot(1,3,1)
plt.imshow(img_2bit, cmap='gray')
plt.title('2 Bit (4 Level)')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(img_4bit, cmap='gray')
plt.title('4 Bit (16 Level)')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(img_8bit, cmap='gray')
plt.title('8 Bit (256 Level)')
plt.axis('off')

plt.show()
```
